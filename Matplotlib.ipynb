### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np,
np.random.seed(3),  
x = 3 + np.random.normal(0, 2, 50), 
y = 3 + np.random.normal(0, 2, len(x)), 
Note: Also add title, xlabel, and ylabel to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
np.random.seed(10)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("X axes")
plt.ylabel("Y axes")
plt.title("Scatter plot of random data collection")
plt.scatter(x,y)
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

"""The subplot () function takes three arguments that describes the layout of the figure. 
The layout is organized in rows and columns, which are represented by the first and second argument. 
The third argument represents the index of the current plot. 
the figure has 1 row, 2 columns, and this plot is the first plot."""

```python
# Line 1
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(1,2,1)
plt.plot(x,y)

# line 2
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(1,2,1)
plt.plot(x,y)

# line 3: 
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(1,2,1)
plt.plot(x,y)

# line 4: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(1,2,1)
plt.plot(x,y)

plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
# -->The bar plots can be plotted horizontally or vertically.
# -->which is data that can be divided into distinct categories. 
# -->In a bar plot, each category is represented by a separate bar, and 
# -->the height of the bar represents the frequency or count of data points in that category.
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
```

```python
# Horizontal bar plot
plt.barh(company,profit)
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
# -->A Box Plot is also known as Whisker plot is created to display the summary of 
#    the set of data values having properties like minimum, first quartile, median, third quartile and maximum.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(10,4))
plt.boxplot(box1)
plt.boxplot(box2)
plt.show()
```
