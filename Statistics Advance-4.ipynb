#### Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test

Z-Test:

> Population Standard Deviation Known:
    A z-test is used when the population standard deviation is known.
> Example Scenario:
    Imagine you are conducting a study on the average height of adult males in a city, 
    and you know the standard deviation of the population heights. You collect a sample of 
    30 adult males and want to test whether the sample mean height is significantly different 
    from the known population mean.

T-Test:

> Population Standard Deviation Unknown:
    A t-test is used when the population standard deviation is unknown and needs to be estimated from the sample data.
> Example Scenario:
    Suppose you are studying the effectiveness of a new teaching method in a school and you want to
    compare the test scores of students who were taught using the new method versus those taught using 
    the traditional method. Since you don't know the population standard deviation of test scores, you use a t-test.

#### Q2: Differentiate between one-tailed and two-tailed tests.

> One-Tailed Test:

    In a one-tailed (or one-sided) test, the critical region is located in only one tail of the distribution of the test
    statistic.
    This type of test is used when you are specifically interested in whether the sample parameter is significantly greater than
    or less than a certain value, but not both.
    The hypothesis statements are often written as either "greater than" or "less than."
    It is more powerful than a two-tailed test when the effect is expected to be in a specific direction.

    Example of One-Tailed Test:
    Hypothesis:
    H0 (null hypothesis): The average score is equal to 70.
    H1 (alternative hypothesis): The average score is greater than 70.
    Critical region: Only the upper tail of the distribution is considered.

> Two-Tailed Test:

    In a two-tailed (or two-sided) test, the critical region is divided between both tails of the distribution of the test
    statistic.
    This type of test is used when you are interested in whether the sample parameter is significantly different from a certain
    value, either greater or less than.
    The hypothesis statements are often written as "not equal to."
    It is more conservative than a one-tailed test as it considers both extremes.
    
    Example of Two-Tailed Test:

    Hypothesis:
    H0 (null hypothesis): The average score is equal to 50.
    H1 (alternative hypothesis): The average score is not equal to 50.
    Critical region: Both the upper and lower tails of the distribution are considered.

#### Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

> Type I Error:

    Definition: Type I error occurs when you reject a true null hypothesis. In other words, you conclude that there is a 
    significant effect when, in reality, there is no effect.
    Symbolically: Denoted as α (alpha), the probability of committing a Type I error is the significance level of the test.
    Example Scenario:
    Suppose you are conducting a medical test to determine if a patient has a rare disease. The null hypothesis (H0) is that 
    the patient is healthy. A Type I error would occur if you incorrectly reject the null hypothesis and conclude that the 
    patient has the disease, while, in fact, the patient is healthy.
> Type II Error:

    Definition: Type II error occurs when you fail to reject a false null hypothesis. In other words, you conclude that there 
    is no significant effect when, in reality, there is an effect.
    Symbolically: Denoted as β (beta), the probability of committing a Type II error is related to the power of the test.
    Example Scenario:
    Using the same medical test example, let's say the null hypothesis (H0) is that the patient is healthy when, in fact, they
    have the disease. A Type II error would occur if you incorrectly fail to reject the null hypothesis and conclude that the
    patient is healthy, even though they actually have the disease.

#### Q4:  Explain Bayes's theorem with an example.

Bayes’s theorem is a mathematical formula that allows us to calculate the probability of an event based on some prior or
related information. It is often used in statistics, data analysis, and machine learning to update our beliefs or hypotheses
based on new evidence or data.
The general formula for Bayes’s theorem is:
P(A∣B)=P(B∣A)*P(A)/P(B)​
where:

P(A∣B) is the conditional probability of event A given that event B has occurred, also called the posterior probability.
P(B∣A) is the conditional probability of event B given that event A has occurred, also called the likelihood.
P(A) is the probability of event A without any condition, also called the prior probability.
P(B) is the probability of event B without any condition, also called the marginal probability or the evidence.

To illustrate Bayes’s theorem with an example, let us consider a simple scenario of medical diagnosis. Suppose there is a 
disease that affects 1% of the population, and there is a test that can detect the disease with 99% accuracy (meaning that 
it has a 1% false positive rate and a 1% false negative rate). If a person takes the test and gets a positive result, what is 
the probability that they actually have the disease?
To answer this question, we can use Bayes’s theorem as follows:

Let A be the event that the person has the disease, and B be the event that the test result is positive.
We are given that P(A)=0.01, P(B∣A)=0.99, and P(B∣¬A)=0.01, where ¬A is the complement of A (meaning that the person does not 
have the disease).
We want to find P(A∣B), the probability that the person has the disease given that the test result is positive.
Using Bayes’s theorem, we get:

P(A∣B)=P(B∣A)P(A)​/P(B)

To calculate P(B), we can use the law of total probability, which states that the probability of an event is equal to the 
sum of the probabilities of the event conditioned on each possible outcome of another event. In this case, we have:

P(B)=P(B∣A)P(A)+P(B∣¬A)P(¬A)

Substituting the given values, we get:

P(B)=0.0198

Therefore, the posterior probability is:

P(A∣B)​≈0.5

This means that the probability that the person has the disease given that the test result is positive is about 50%, which 
is much lower than the accuracy of the test. This is because the disease is very rare in the population, so the prior 
probability of having the disease is very low. The test result only slightly increases the likelihood of having the disease, 
but not enough to overcome the low prior probability

(the probability of an event given new evidence) to the prior probability (our initial belief) and the likelihood.
The general form of Bayes’ Theorem is:
P(A∣B)=P(B∣A)⋅P(A)/P(B)​

where:
P(A∣B) is the posterior probability (the probability of event A given evidence B).
P(B∣A) is the likelihood (the probability of evidence B given event A).
P(A) is the prior probability (our initial belief in event A).
P(B) is the evidence (the probability of observing evidence B).
